### This is an exercise with pandas from kaggle_shoppe 

```python
import pandas as pd
from datetime import datetime, timedelta

origin_order = pd.read_csv(r'C:\Users\Tibame\Danny\Data_Mining\class_0817\delivery_orders_march.csv', sep=",")
# make a copy 
order = origin_order.copy()

# take a look
print('rows =', order.shape[0], '\n' 'columns =', order.shape[1])
order.head()
```

### Transform epoch time to datetime 

```python
# transform epoch time to datetime 
def transform_EpochToDatetime(x):
    return pd.Timestamp(x, unit='s').date() + timedelta(hours=8)

order['pick_date'] = order['pick'].apply(transform_EpochToDatetime)
order['1st_date'] = order['1st_deliver_attempt'].apply(transform_EpochToDatetime)
order['2nd_date'] = order['2nd_deliver_attempt'].apply(transform_EpochToDatetime)

order.head()
```

### Transform buyer/seller address --> Metro Manila, Luzon, Visayas, Mindanao 

```python
# cat order['buyaddress']
for i, j in enumerate(order['buyeraddress'].head()):
    print(i, '\t', j, '\n')
```

```python
# Create new columns --> buyer_area, seller_area
# value in [Metro Manila, Luzon, Visayas, Mindanao]
order['buyer_area'] = order['buyeraddress'].apply(lambda x: x.split(' ')[-1])
order['seller_area'] = order['selleraddress'].apply(lambda x: x.split(' ')[-1])

order.loc[order['buyer_area'].str.contains('Manila|manila'), 'buyer_area'] = 'Metro Manila'
order.loc[order['buyer_area'].str.contains('Luzon|luzon'), 'buyer_area'] = 'Luzon'
order.loc[order['buyer_area'].str.contains('Visayas|visayas'), 'buyer_area'] = 'Visayas'
order.loc[order['buyer_area'].str.contains('Mindanao|mindanao'), 'buyer_area'] = 'Mindanao'

order.loc[order['seller_area'].str.contains('Manila|manila'), 'seller_area'] = 'Metro Manila'
order.loc[order['seller_area'].str.contains('Luzon|luzon'), 'seller_area'] = 'Luzon'
order.loc[order['seller_area'].str.contains('Visayas|visayas'), 'seller_area'] = 'Visayas'
order.loc[order['seller_area'].str.contains('Mindanao|mindanao'), 'seller_area'] = 'Mindanao'

# Make sure this orderDate have become what I need
order[['orderid', 'pick_date', '1st_date', '2nd_date', 'buyer_area', 'seller_area']].head()
```

### Find out holidays

```python
holidays_list = [
    '2020-03-01',
    '2020-03-08',
    '2020-03-15',
    '2020-03-22',
    '2020-03-25',
    '2020-03-29',
    '2020-03-30',
    '2020-03-31',
    '2020-04-05'  # if pick_date is 03-31
]

holidays = [datetime.strptime(i, '%Y-%m-%d').date() for i in holidays_list]

order['1st_holidays'] = 0
order['2nd_holidays'] = 0

for holiday in holidays:
    order.loc[(order['pick_date'] <= holiday) & (order['1st_date'] >= holiday), '1st_holidays'] += 1
    order.loc[(order['1st_date'] <= holiday) & (order['2nd_date'] >= holiday), '2nd_holidays'] += 1

# turn order[1st_holidays, 2nd_holidays] into timedelta
order['1st_holidays'] = order['1st_holidays'].apply(lambda x: timedelta(days=x))
order['2nd_holidays'] = order['2nd_holidays'].apply(lambda x: timedelta(days=x))

order[['orderid', 'pick_date', '1st_date', '2nd_date', '1st_holidays', '2nd_holidays', 'buyer_area', 'seller_area']].head()
```

### Deduct holidays

```python
order['final_1st'] = order['1st_date'] - order['pick_date'] - order['1st_holidays']
order['final_2nd'] = order['2nd_date'] - order['1st_date'] - order['2nd_holidays']

order[['orderid', 'pick_date', '1st_date', '2nd_date', '1st_holidays', \
       '2nd_holidays', 'buyer_area', 'seller_area', 'final_1st', 'final_2nd']]\
        .head()
```

### Create the SLA column

```python
order['sla'] = timedelta(days=7)

order.loc[(order['buyer_area'] == 'Metro Manila') & (order['seller_area'] == 'Metro Manila'), 'sla'] = timedelta(days=3)
order.loc[(order['buyer_area'] == 'Metro Manila') & (order['seller_area'] == 'Luzon'), 'sla'] = timedelta(days=5)
order.loc[(order['buyer_area'] == 'Luzon') & (order['seller_area'] == 'Metro Manila'), 'sla'] = timedelta(days=5)
order.loc[(order['buyer_area'] == 'Luzon') & (order['seller_area'] == 'Luzon'), 'sla'] = timedelta(days=5)

order[['orderid', 'pick_date', '1st_date', '2nd_date', '1st_holidays', \
       '2nd_holidays', 'buyer_area', 'seller_area', 'final_1st', 'final_2nd', \
       'sla']]\
        .head()
```

### Submit final data

```python
final_data = order.loc[:, ['orderid', 'final_1st', 'final_2nd', 'sla']]
final_data
```

```python
final_data['is_late'] = 0

final_data.loc[final_data['final_1st'] > final_data['sla'], 'is_late'] = 1
final_data.loc[final_data['final_2nd'] > timedelta(days=3), 'is_late'] = 1

final_data.head(20)
```

```python
# Save final_data to csv
final_data[['orderid', 'is_late']].to_csv('./kaggle_shopee.csv', index=False)
```
